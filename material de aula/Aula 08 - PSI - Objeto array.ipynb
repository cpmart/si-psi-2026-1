# Fundamentos {#fundamentos}

Um **Array** em JavaScript é um objeto especial destinado a armazenar uma **coleção ordenada de dados**, indexados sequencialmente a partir de **zero**.

> **Comparação com Python:** o Array do JS equivale à `list` do Python. Ambos são dinâmicos, aceitam tipos mistos e usam índice inteiro. A diferença está na API de métodos.

## Criando um Array

```javascript
let a = new Array();  // via construtor (evite)
let b = [];           // literal — forma recomendada
```

Use sempre a **notação literal `[]`** — é mais concisa e menos ambígua (o construtor `new Array(3)` cria um array com 3 posições *vazias*, não com o valor `3`).

## Populando

```javascript
// Ao declarar
let misto = ["casa", "rio", 5, "Jales", 12.5];

// Após declarar
let a = [];
a[0] = 3.14159;
a[1] = "prego";
a[2] = true;
a[3] = { nome: "José", cidade: "Jales" };
```

Arrays JS são **heterogêneos**: cada posição pode ter um tipo diferente, inclusive objetos e outros arrays.

## Acessando elementos

````javascript
// Arrays com tipos mistos e objetos embutidos
let misto = ["casa", "rio", 5, "Jales", 12.5];
console.log(misto[0]);   // "casa"
console.log(misto[4]);   // 12.5

let outroArray = ["casa", "rio", { x: 14, y: "mar" }];
console.log(outroArray[2].x);      // 14  — notação ponto
console.log(outroArray[2]["x"]);   // 14  — notação colchete
console.log(outroArray[2].y);      // "mar"

// Array criado aos poucos
let a = [];
a[0] = 3.14159;
a[1] = "prego";
a[2] = true;
a[3] = { nome: "José", cidade: "Jales" };
console.log(a[0]);          // 3.14159
console.log(a[3].cidade);   // "Jales"
````

# Propriedades do Construtor {#propriedades-do-construtor}

**Propriedade >> Descrição**

- `length`: Número de elementos do array 
- `constructor`: Referência à função construtora 
- `prototype`: Permite adicionar propriedades/métodos a todos os arrays 

A propriedade mais usada no dia a dia é `length`:

````javascript
let arr = [1, 2, 3, "a", "b", "c", { nome: "Maria", cidade: "Jales" }];
console.log(arr.length);  // 7 — conta objetos como 1 elemento

// Iterar pelo índice com length (forma clássica)
for (let i = 0; i < arr.length; i++) {
    console.log(i, arr[i]);
}
````


# Imutabilidade vs. Mutabilidade {#imutabilidade-mutabilidade}

Um aspecto fundamental — especialmente em código moderno — é saber **se um método altera o array original** ou **retorna um novo array**.

>**Categoria >> Métodos principais >> O array original...**

**Imutáveis** 

- `slice`, `concat`, `map`, `filter`, `reduce`, `find`, `every`, `some`, `indexOf`, `includes`, `join`, `flat`, `flatMap`  
- Não é alterado 

**Mutáveis**  

- `push`, `pop`, `shift`, `unshift`, `splice`, `reverse`, `sort`, `fill`, `copyWithin` 
- É alterado diretamente 

> **Boas práticas modernas:** prefira métodos imutáveis sempre que possível. Código funcional (usando `map`, `filter`, `reduce`) é mais previsível, testável e compatível com frameworks como React.

> **Nota ES2023:** `Array.prototype.toSorted()`, `toReversed()` e `toSpliced()` foram adicionados ao padrão justamente para oferecer versões *imutáveis* de `sort`, `reverse` e `splice`. São suportados nos navegadores modernos.

## Métodos Imutáveis

### `concat()` — une arrays sem modificar os originais

````javascript
let arr = [1, 2, "a", "b"];
let a = arr.concat("m", 32);
console.log(a);    // [1, 2, "a", "b", "m", 32]
console.log(arr);  // [1, 2, "a", "b"]  — original intacto
````

### `slice(inicio, fim)` — extrai sub-array

- Inclui o índice `inicio`; **exclui** o índice `fim`.
- Índices negativos contam a partir do final (`-1` = último elemento).
- Sem `fim`: vai até o final do array.

````javascript
let arr = [1, 2, 3, 4, "a", "b", "c"];
//          0  1  2  3   4    5    6
//         -7 -6 -5 -4  -3   -2   -1

console.log(arr.slice(2, 5));    // [3, 4, "a"]
console.log(arr.slice(3));       // [4, "a", "b", "c"]
console.log(arr.slice(-5, 6));   // [3, 4, "a", "b"]
console.log(arr.slice(-6, -4));  // [2, 3]
console.log(arr);                // original intacto
````

### `join(separador)` — converte array em string

````javascript
let nomes = ["Alice", "Bob", "Carlos", "Diana"];
console.log(nomes.join("; "));
// Alice; Bob; Carlos; Diana

// Construindo frase natural com pop + join
let nomesParaFrase = [...nomes];           // cópia para não alterar original
let ultimoNome = nomesParaFrase.pop();
let frase = nomesParaFrase.join(", ") + " e " + ultimoNome;
console.log(frase);
// Alice, Bob, Carlos e Diana
````

## Métodos Mutáveis

### `push()` e `pop()` — manipulam o *final* do array

**Método >> Ação >> Retorna**

- `push(...items)` >> Adiciona ao final >> novo `length` 
- `pop()` >> Remove do final >> elemento removido 

````javascript
let pilha = [];
pilha.push("Dom Quixote");
pilha.push("1984");
pilha.push("O Pequeno Príncipe");
console.log(pilha);
// ["Dom Quixote", "1984", "O Pequeno Príncipe"]

let removido = pilha.pop();
console.log(removido);  // "O Pequeno Príncipe"
console.log(pilha);     // ["Dom Quixote", "1984"]
````

### `shift()` e `unshift()` — manipulam o *início* do array

**Método >> Ação >> Retorna**

- `unshift(...items)` >> Adiciona ao início >> novo `length` 
- `shift()` >> Remove do início >> elemento removido 


> **Desempenho:** `shift` e `unshift` são mais lentos que `pop` e `push` pois precisam reindexar todos os elementos. Para filas de alto desempenho, considere usar uma estrutura dedicada.

````javascript
let fila = ["João", "Maria", "Ana"];

let atendido = fila.shift();   // remove o primeiro
console.log(atendido);         // "João"
console.log(fila);             // ["Maria", "Ana"]

fila.unshift("Julia");         // emergência entra na frente
console.log(fila);             // ["Julia", "Maria", "Ana"]
````

### `splice(pos, qtd, ...itens)` — remove, insere ou substitui

É o método mutável mais versátil:

```
splice(posição, quantidadeARemover, item1, item2, ...)
```

- **Retorna** array com os elementos *removidos*.
- Com `qtd = 0`: apenas insere, sem remover.

````javascript
let arr = [1, 2, 3, 4, "a", "b", "c"];

// 1) Remover 1 elemento a partir do índice 1
console.log("splice(1,1) ->", arr.splice(1, 1)); // [2]
console.log("arr:", arr);  // [1, 3, 4, 'a', 'b', 'c']

// 2) Inserir sem remover (qtd = 0)
arr.splice(2, 0, "X", "Y");
console.log("arr após inserção:", arr);
// [1, 3, "X", "Y", 4, 'a', 'b', 'c']

// 3) Substituir (remover 1 e inserir 1)
arr.splice(4, 1, "Z");
console.log("arr após substituição:", arr);
````

### `reverse()` — inverte no lugar

> **ES2023:** use `toReversed()` para obter um novo array invertido sem alterar o original.

````javascript
let nums = [1, 2, 3, 4, 5];
nums.reverse();
console.log(nums);  // [5, 4, 3, 2, 1]

// Verificador de palíndromo
function ehPalindromo(palavra) {
    let invertida = palavra.split("").reverse().join("");
    return palavra === invertida;
}
console.log(ehPalindromo("radar"));   // true
console.log(ehPalindromo("hello"));   // false
console.log(ehPalindromo("arara"));   // true
````

# Métodos de Busca e Verificação

## `indexOf()` e `lastIndexOf()` — retornam posição ou `-1`

````javascript
let arr = [1, 2, "a", "b", "2", "c", 1, "a", "45"];

console.log(arr.indexOf("b"));   // 3
console.log(arr.indexOf("a"));   // 2  (primeira ocorrência)
console.log(arr.indexOf(23));    // -1 (não encontrado)
console.log(arr.indexOf(2));     // 1  (tipo importa: 2 !== "2")
console.log(arr.indexOf(45));    // -1 (45 !== "45")

// Busca a partir de um índice
let frutas = ["Banana", "Laranja", "Maçã", "Manga", "Maçã"];
console.log(frutas.indexOf("Maçã"));     // 2
console.log(frutas.indexOf("Maçã", 3));  // 4 (busca a partir do índice 3)

// lastIndexOf: última ocorrência
console.log(frutas.lastIndexOf("Maçã")); // 4
````

## `includes()` — verificação booleana (mais moderno)

> **Diferença chave em relação a `indexOf`:** `includes` trata `NaN` corretamente (`indexOf(NaN)` sempre retorna `-1`, mesmo que o array contenha `NaN`).

````javascript
const convidados = ["Alice", "Bob", "Charlie", "Dana"];
console.log(convidados.includes("Charlie"));  // true
console.log(convidados.includes("Eve"));      // false

// Caso especial: NaN
const arrayComNaN = [NaN, 2, 3];
console.log(arrayComNaN.includes(NaN));  // true
console.log(arrayComNaN.indexOf(NaN));   // -1  (limitação de indexOf)
````

## `find()` e `findIndex()` — busca com predicado

- `find()`: retorna o **primeiro elemento** que satisfaz a condição (ou `undefined`).
- `findIndex()`: retorna o **índice** do primeiro elemento que satisfaz a condição (ou `-1`).
- `findLast()` / `findLastIndex()`: versões que buscam do **final para o início** (ES2023).

````javascript
let livros = [
    { titulo: "O Pequeno Príncipe", paginas: 96 },
    { titulo: "Dom Quixote",        paginas: 1072 },
    { titulo: "O Hobbit",           paginas: 295 }
];

// find — retorna o objeto
let livroGrande = livros.find(livro => livro.paginas > 300);
console.log(livroGrande);
// { titulo: "Dom Quixote", paginas: 1072 }

// findIndex — retorna o índice
let idx = livros.findIndex(livro => livro.paginas > 300);
console.log(idx);  // 1

// Se nenhum satisfaz a condição
let enorme = livros.find(livro => livro.paginas > 5000);
console.log(enorme);  // undefined
````

# Métodos de Iteração e Transformação

Estes são os métodos mais importantes da programação funcional com arrays. Todos recebem uma **função callback** com a assinatura `(elemento, índice, array)`.

## `forEach()` — itera sem retornar valor

> Equivalente ao `for elem in lista:` do Python, mas sem `break`/`continue`. Para interromper, use `for...of` ou `some`.

````javascript
let numeros = [1, 2, 3, 4, 5];

numeros.forEach(n => console.log(n * 2));
// 2, 4, 6, 8, 10

// Com todos os parâmetros do callback
numeros.forEach((numero, indice, array) => {
    console.log(`[${indice}] = ${numero} | total: ${array.length}`);
});
````

## `map()` — transforma cada elemento, retorna novo array

> Equivalente a `[f(x) for x in lista]` em Python (list comprehension).

````javascript
let celsius = [0, 10, 20, 30, 40];

let fahrenheit = celsius.map(c => (c * 9 / 5) + 32);
console.log(fahrenheit);
// [32, 50, 68, 86, 104]

// Transformando array de objetos
let usuarios = [
    { nome: "Ana",  ativo: true },
    { nome: "João", ativo: false }
];
let nomes = usuarios.map(u => u.nome);
console.log(nomes);  // ["Ana", "João"]
````

## `filter()` — seleciona elementos que satisfazem uma condição

> Equivalente a `[x for x in lista if condicao(x)]` em Python.

````javascript
let pessoas = [
    { nome: "Ana",     idade: 22 },
    { nome: "João",    idade: 17 },
    { nome: "Mariana", idade: 18 },
    { nome: "Pedro",   idade: 15 }
];

let adultos = pessoas.filter(p => p.idade >= 18);
console.log(adultos);
// [{nome:"Ana",idade:22}, {nome:"Mariana",idade:18}]

// Filtrando por índice
let arr = [21, 3, 18, "a", 290, "b", 7];
let primeiros4 = arr.filter((elem, ind) => ind < 4);
console.log(primeiros4);  // [21, 3, 18, "a"]
````

## `reduce()` e `reduceRight()` — acumula valores

```javascript
array.reduce((acumulador, valorAtual, indice, array) => novoAcumulador, valorInicial)
```

- `reduceRight`: percorre do **último para o primeiro** elemento.

````javascript
let numeros = [1, 2, 3, 4, 5];

// Soma
let soma = numeros.reduce((acc, val) => acc + val, 0);
console.log(soma);  // 15

// Maior valor (sem Math.max)
let maior = [8, 5, 1, 9, 20, 2].reduce((a, b) => (a > b ? a : b));
console.log(maior);  // 20

// Contando ocorrências — uso avançado
let frutas = ["maçã", "banana", "maçã", "laranja", "banana", "maçã"];
let contagem = frutas.reduce((acc, fruta) => {
    acc[fruta] = (acc[fruta] || 0) + 1;
    return acc;
}, {});
console.log(contagem);
// { maçã: 3, banana: 2, laranja: 1 }

// reduceRight
let transacoes = [200, -50, 300, -100, 50];
let saldo = transacoes.reduceRight((acc, val) => acc + val, 0);
console.log(saldo);  // 400
````

## `every()` e `some()` — testes booleanos

**Método retorna `true` quando...** 

- `every(fn)`: **Todos** os elementos satisfazem `fn`  
- `some(fn)`: **Pelo menos um** satisfaz `fn` 

````javascript
let arr = [21, 3, 18, 290];
console.log(arr.every(e => typeof e === "number"));  // true
console.log(arr.every(e => e > 10));                 // false (3 não é > 10)

let misto = [21, 3, 18, "a", 290];
console.log(misto.some(e => typeof e === "string")); // true

let numeros = [3, 7, 11, 5, 2];
console.log(numeros.some(n => n > 10));              // true (11 > 10)
````

## `flat()` e `flatMap()` — achata arrays aninhados (ES2019)


````javascript
// flat — achata um nível por padrão
let aninhado = [1, [2, 3], [4, [5, 6]]];
console.log(aninhado.flat());     // [1, 2, 3, 4, [5, 6]]
console.log(aninhado.flat(2));    // [1, 2, 3, 4, 5, 6]
console.log(aninhado.flat(Infinity)); // achata qualquer profundidade

// flatMap — equivale a map() seguido de flat(1)
let frases = ["Olá mundo", "JavaScript é legal"];
let palavras = frases.flatMap(frase => frase.split(" "));
console.log(palavras);
// ["Olá", "mundo", "JavaScript", "é", "legal"]
````

# Métodos de Criação

## `Array.from()` — cria array a partir de iterável ou array-like

````javascript
// String → Array
console.log(Array.from("Hello"));
// ["H", "e", "l", "l", "o"]

// Set → Array (útil para remover duplicatas)
const set = new Set([1, 2, 3, 2, 1]);
console.log(Array.from(set));  // [1, 2, 3]

// Com função de mapeamento (segundo argumento)
let sequencia = Array.from({ length: 5 }, (_, i) => i + 1);
console.log(sequencia);  // [1, 2, 3, 4, 5]

let quadrados = Array.from({ length: 4 }, (_, i) => i * i);
console.log(quadrados);  // [0, 1, 4, 9]
````

## `Array.of()` — cria array com os argumentos fornecidos

> **Por que existe?** `new Array(3)` cria `[empty × 3]`, não `[3]`. `Array.of(3)` resolve essa ambiguidade criando `[3]`.

````javascript
console.log(Array.of(1, 2, 3));              // [1, 2, 3]
console.log(Array.of("Hello", "World"));      // ["Hello", "World"]
console.log(Array.of(1, "Apple", { a: 1 })); // [1, "Apple", {a:1}]

// Diferença crítica com new Array
console.log(new Array(3));    // [ <3 empty slots> ] — armadilha!
console.log(Array.of(3));     // [3]
````

## `fill(valor, inicio, fim)` — preenche com um valor

````javascript
// Criar array de zeros
let zeros = new Array(5).fill(0);
console.log(zeros);  // [0, 0, 0, 0, 0]

// Preencher faixa específica
let arr = [1, 2, 3, 4, 5];
arr.fill(9, 1, 4);  // índices 1 a 3
console.log(arr);   // [1, 9, 9, 9, 5]
````

## `split()` — converte string em array (método de `String`)

> **Atenção:** `split` é um método de **String**, não de Array. Está aqui pois é frequentemente usado em conjunto com arrays.

````javascript
let data = "25/04/2025";
let partes = data.split("/");
console.log(partes);  // ["25", "04", "2025"]

let nomes = "Alice, Bob, Carlos, Diana";
console.log(nomes.split(", "));
// ["Alice", "Bob", "Carlos", "Diana"]

// Limitando o número de partes
let texto = "uma, duas, três, quatro";
console.log(texto.split(", ", 3));
// ["uma", "duas", "três"]
````

## `at()` — acesso por índice com suporte a negativos (ES2022)

````javascript
const arr = [3, 7, 2, 5, 4];

console.log(arr.at(0));   // 3  — mesmo que arr[0]
console.log(arr.at(1));   // 7
console.log(arr.at(-1));  // 4  — último elemento!
console.log(arr.at(-3));  // 2

// Forma tradicional (verbosa) para pegar o último
console.log(arr[arr.length - 1]);  // 4 — equivalente a at(-1)

// Também funciona em strings
let saudacao = "Olá, mundo!";
console.log(saudacao.at(-1));  // "!"
````

# Ordenação

## `sort()` — ordena no lugar (mutável)

> **Armadilha importante:** sem função comparadora, `sort` converte elementos para **string** antes de comparar. Isso faz `[10, 5, 100]` ser ordenado como `[10, 100, 5]` (ordem lexicográfica)! Sempre forneça a função comparadora para números.

````javascript
// ARMADILHA: sort sem comparador
let errado = [10, 5, 100, 50, 1];
errado.sort();
console.log(errado);  // [1, 10, 100, 5, 50]  — ERRADO para números!

// Correto: com função comparadora
let numeros = [10, 5, 100, 50, 1];
numeros.sort((a, b) => a - b);     // crescente
console.log(numeros);  // [1, 5, 10, 50, 100]

numeros.sort((a, b) => b - a);     // decrescente
console.log(numeros);  // [100, 50, 10, 5, 1]

// Strings — ordem alfabética
let frutas = ["banana", "maçã", "uva", "manga"];
frutas.sort();
console.log(frutas);  // ["banana", "manga", "maçã", "uva"]

// Strings — ignorando maiúsculas/minúsculas com localeCompare
let itens = ["banana", "Maçã", "Uva", "manga"];
itens.sort((a, b) => a.toLowerCase().localeCompare(b.toLowerCase()));
console.log(itens);  // ["banana", "Maçã", "manga", "Uva"]

// Ordenando array de objetos por propriedade
let funcionarios = [
    { nome: "João",    salario: 3000 },
    { nome: "Maria",   salario: 3200 },
    { nome: "Juliana", salario: 2800 },
    { nome: "Pedro",   salario: 4000 },
    { nome: "Mário",   salario: 2750 }
];

// Por nome (alfabético)
funcionarios.sort((a, b) => a.nome.localeCompare(b.nome));
console.log(funcionarios.map(f => f.nome));
// ["João", "Juliana", "Maria", "Mário", "Pedro"]

// Por salário (decrescente)
funcionarios.sort((a, b) => b.salario - a.salario);
console.log(funcionarios.map(f => `${f.nome}: ${f.salario}`));
````

## `toSorted()` — versão imutável de `sort` (ES2023)

> Disponível nos navegadores modernos e Node.js 20+.

````javascript
const original = [3, 1, 4, 1, 5, 9];
const ordenado = original.toSorted((a, b) => a - b);
console.log(original);  // [3, 1, 4, 1, 5, 9]  — intacto!
console.log(ordenado);  // [1, 1, 3, 4, 5, 9]
````

# Desestruturação de Array (ES6)

A **desestruturação** permite extrair valores de arrays e atribuí-los a variáveis de forma concisa — análoga ao *unpacking* do Python (`a, b = [1, 2]`).

````javascript
// Equivalente Python: a, b = ['x', 'y']
let [x, y] = ["a", "b"];
console.log(x);  // "a"
console.log(y);  // "b"

// Troca de valores sem variável temporária
[x, y] = [y, x];

console.log(x);  // "b"
console.log(y);  // "a"

// Desestruturando array existente
let frutas = ["maçã", "banana", "laranja"];
let [fruta1, fruta2, fruta3] = frutas;
console.log(fruta1, fruta2, fruta3);

// Ignorando elementos com vírgula
let [primeiro,  , terceiro] = frutas;
console.log(primeiro, terceiro);  // "maçã" "laranja"

// Rest element — captura o restante
let [cabeca, ...cauda] = [1, 2, 3, 4, 5];
console.log(cabeca);  // 1
console.log(cauda);   // [2, 3, 4, 5]

// Desestruturando retorno de função
function coordenadas() { return [45.123, -93.456]; }
let [lat, lon] = coordenadas();
console.log(`Lat: ${lat}, Lon: ${lon}`);
````

# Operador Spread (`...`)

Introduzido no ES6, o operador **spread** (`...`) expande um iterável em posições individuais. É complementar à desestruturação.

## Usos principais com arrays:

````javascript
// 1) Clonar array (cópia rasa)
let original = [1, 2, 3];
let copia = original;
copia.push(4);
console.log(original);  // [1, 2, 3] — intacto
console.log(copia);     // [1, 2, 3, 4]

// 2) Combinar (merge) arrays
let a1 = [0, 1, 2];
let a2 = [3, 4, 5];
let combinado = [...a1, ...a2];
console.log(combinado);  // [0, 1, 2, 3, 4, 5]

// 3) Inserir em posição
let partes = ["shoulders", "knees"];
let musica = ["head", ...partes, "and", "toes"];
console.log(musica);
// ["head", "shoulders", "knees", "and", "toes"]

// 4) String → Array
let letras = [... "ABC"];
console.log(letras);  // ["A", "B", "C"]

// 5) Set → Array (remove duplicatas)
let semDup = [...new Set([1, 2, 3, 1, 2])];
console.log(semDup);  // [1, 2, 3]

// 6) Spread em objetos (ES2018)
let obj1 = { foo: "bar", x: 42 };
let obj2 = { foo: "baz", y: 15 };
let clone = { ...obj1 };
let merged = { ...obj1, ...obj2 };
console.log(clone);   // { foo: "bar", x: 42 }
console.log(merged);  // { foo: "baz", x: 42, y: 15 }  — foo de obj2 vence
````

# Arrays Bidimensionais (Matrizes)

JavaScript não possui um tipo nativo de matriz — usa-se **arrays de arrays**.

````javascript
// Forma 1: literal
let matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
];
console.log(matriz[1][2]);  // 6 — linha 1, coluna 2

// Forma 2: programática
let m = new Array(3);
//  m = [ , , ]
for (let i = 0; i < m.length; i++) {
    m[i] = new Array(3).fill(0);  // preenchida com zeros
}
/*
[
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
];
*/
m[0][0] = 1;
m[1][1] = 5;
m[2][2] = 9;
/*
[
    [1, 0, 0],
    [0, 5, 0],
    [0, 0, 9]
];
*/

console.log(m);

// Adicionando linha
matriz.push([10, 11, 12]);
console.log(matriz.length);  // 4 linhas

// Percorrendo com for...of
for (let linha of matriz) {
    console.log(linha.join(" | "));
}
````

# Objeto Set

O **Set** armazena **valores únicos** de qualquer tipo. Diferentemente de um array, não admite duplicatas. É análogo ao `set` do Python.

## Operações e  Métodos 

- Adicionar: `add(valor)` 
- Verificar: `has(valor)` 
- Remover: `delete(valor)` 
- Limpar tudo:  `clear()` 
- Tamanho:  `.size` (não `.length`!) 

````javascript
let s = new Set();
s.add(1);
s.add(5);
s.add(5);       // ignorado — já existe
s.add("texto");
let o = { a: 1, b: 2 };
s.add(o);
s.add({ a: 1, b: 2 });  // objeto diferente → aceito (referência distinta)
console.log(s.size);  // 5

console.log(s.has(1));                        // true
console.log(s.has(3));                        // false
console.log(s.has(Math.sqrt(25)));            // true (sqrt(25) = 5)
console.log(s.has("Texto".toLowerCase()));    // true

s.delete(5);
console.log(s.size);  // 4

// Convertendo entre Array e Set
let myArray = ["v1", "v2", "v3"];
let mySet = new Set(myArray);
console.log(mySet.has("v1"));         // true
console.log([...mySet]);              // ["v1", "v2", "v3"]

// Removendo duplicatas de array — uso mais comum
const nums = [2, 3, 4, 4, 2, 3, 3, 4, 5, 6, 7, 32];
console.log([...new Set(nums)]);
// [2, 3, 4, 5, 6, 7, 32]
````

# Exemplos Integrados

## Encadeamento de métodos (method chaining)

````javascript
// Alunos aprovados com mais de 20 anos
const estudantes = [
    { nome: "João",   idade: 50, nota: 5.9 },
    { nome: "Pietra", idade: 21, nota: 10.0 },
    { nome: "Diego",  idade: 18, nota: 9.0 }
];

let resultado = estudantes
    .filter(a => a.idade > 20 && a.nota > 6)
    .map(a => `${a.nome} - Aprovado`);

console.log(resultado);  // ["Pietra - Aprovado"]

// Maior número menor que 10
const nums = [5, 12, 8, 14, 6];
const maiorMenorQueDez = nums
    .filter(n => n < 10)
    .reduce((a, b) => a > b ? a : b);
console.log(maiorMenorQueDez);  // 8
````

````javascript
// Ordenação e extração
const produtos = [
    { nome: "ProdutoA", preco: 30 },
    { nome: "ProdutoB", preco: 20 },
    { nome: "ProdutoC", preco: 40 }
];

let nomesOrdenados = [...produtos]
    .sort((a, b) => b.preco - a.preco)
    .map(p => p.nome);

console.log(nomesOrdenados);  // ["ProdutoC", "ProdutoA", "ProdutoB"]
````

````javascript
// Combinando arrays de objetos (JOIN de duas fontes)
let usuarios = [
    { id: 1, nome: "Alice" },
    { id: 2, nome: "Bob" }
];
let enderecos = [
    { userId: 1, endereco: "Rua 123, Cidade A" },
    { userId: 2, endereco: "Rua 456, Cidade B" }
];

let usuariosComEndereco = usuarios.map(u => ({
    ...u,
    endereco: enderecos.find(e => e.userId === u.id).endereco
}));

console.log(usuariosComEndereco);
// [{ id:1, nome:"Alice", endereco:"Rua 123, Cidade A" }, ...]
````

## Exercício 1 — Lista de Tarefas (CRUD com Array)

Implemente um sistema de lista de tarefas com as funções abaixo. Use um array de objetos `{ id, descricao, concluida }`:

- `adicionarTarefa(descricao)` — adiciona ao final

## Exercício 4 — Quiz Interativo

## Exercício 5 — Gráfico de Barras Dinâmico

Crie um array de valores e gere uma visualização em texto (barras de `#`) proporcionais aos valores. Atualize os dados e regenere o gráfico.

**Exemplo de estrutura para o Exercício 1:**

````javascript
// Exercício 1 — esqueleto para desenvolvimento
let tarefas = [];
let contadorId = 1;

function adicionarTarefa(descricao) {
    tarefas.push({ id: contadorId++, descricao, concluida: false });
}

function removerTarefa(id) {
    tarefas = tarefas.filter(t => t.id !== id);
}

function concluirTarefa(id) {
    const tarefa = tarefas.find(t => t.id === id);
    if (tarefa) tarefa.concluida = true;
}

function listarPendentes() {
    return tarefas.filter(t => !t.concluida);
}

// Teste
adicionarTarefa("Estudar JavaScript");
adicionarTarefa("Fazer exercícios");
adicionarTarefa("Revisar anotações");
concluirTarefa(2);

console.log("Todas:", tarefas);
console.log("Pendentes:", listarPendentes());
````

# Resumo dos Métodos de Arrays (Sem Tabela)

## Métodos Mutáveis (alteram o array original)

### Manipulação básica
- `push()` → adiciona no final  
- `pop()` → remove do final  
- `unshift()` → adiciona no início  
- `shift()` → remove do início  

### Manipulação por posição
- `splice()` → remove, insere ou substitui elementos  

### Ordenação e organização
- `reverse()` → inverte a ordem  
- `sort()` → ordena (use comparador para números!)  

### Preenchimento
- `fill()` → preenche o array com um valor  



## Métodos Imutáveis (não alteram o original)

### Criação e cópia
- `concat()` → une arrays  
- `slice()` → extrai parte do array  

### Conversão
- `join()` → transforma array em string  

### Busca
- `indexOf()` / `lastIndexOf()` → retorna posição ou -1  
- `includes()` → retorna true/false  

### Busca com condição
- `find()` → retorna o elemento  
- `findIndex()` → retorna o índice  

### Testes lógicos
- `every()` → todos satisfazem?  
- `some()` → algum satisfaz?  

### Iteração
- `forEach()` → percorre (sem retorno)  

### Transformação
- `map()` → gera novo array transformado  
- `filter()` → filtra elementos  

### Acumulação
- `reduce()` / `reduceRight()` → gera valor único  

### Arrays aninhados
- `flat()` → achata níveis  
- `flatMap()` → map + flat  

### Acesso
- `at()` → acessa índice (aceita negativo)  

### Versões modernas (ES2023)
- `toSorted()` → ordena sem alterar original  
- `toReversed()` → inverte sem alterar original  


## Métodos de criação

- `Array.from()` → cria array a partir de iterável  
- `Array.of()` → cria array a partir de valores
