- `curso`

## Parte 2 — Exercícios Intermediários

Objetivo: trabalhar com métodos dentro de objetos.

### Exercício 6 — Método simples

Crie um objeto chamado **pessoa** com:

- `nome`
- `idade`

Adicione um método chamado **`apresentar`** que mostre:

```text
Olá, meu nome é NOME e tenho IDADE anos.
```

```text
Olá, meu nome é Ana e tenho 21 anos.
```

### Exercício 7 — Calculando desconto

Crie um objeto chamado **produto** com:

- `nome`
- `preco`

Adicione um método chamado **`calcularDesconto`** que receba um percentual e retorne o preço com desconto. Se não for passado valor do desconto, utilize 10% como default.

**Exemplo:**

```javascript
produto.calcularDesconto(10)
```

Se o preço for `100`, o resultado deve ser:

```text
90
```

### Exercício 8 — Conta bancária

Crie um objeto chamado **conta** com:

- `titular`
- `saldo`

Adicione um método chamado **`depositar`** que recebe um valor e soma ao saldo.

Teste:

```javascript
conta.depositar(200)
```

Depois mostre o saldo atualizado.

### Exercício 9 — Carrinho de compras simples

Crie um objeto chamado **carrinho** com:

- `produto`

```text
preco * quantidade
```

### Exercício 10 — Controle de estoque

Crie um objeto chamado **produto** com:

- `nome`
- `estoque`

Adicione um método chamado **`vender`** que receba uma quantidade e diminua do estoque. Depois adicione um método chamado **`comprar`** e amente o estoque.

**Exemplo:**

```text
estoque inicial: 10
venda: 3
estoque final: 7
compra: 2
estoque final: 9
```
