\section*{Lista de Exercícios — Objetos Avançados e Modelagem em JavaScript}

# Aula 03 — Modelagem

Conteúdos trabalhados:

- função construtora
- classes
- herança


## Nível Básico

### Exercício 1 — Função construtora `ContaStreaming`

Crie uma função construtora chamada `ContaStreaming` com as propriedades:

- `usuario`
- `plano`

Depois, crie dois objetos diferentes usando `new`.

### Exercício 2 — Método na construtora

Na função construtora `ContaStreaming`, adicione um método chamado `mostrarPlano()` que exiba uma mensagem como:

`O usuário X está no plano Y.`

### Exercício 3 — Classe `CanalVideo`

Crie uma classe chamada `CanalVideo` com:

- `nome`
- `inscritos`

Use o `constructor` para receber esses valores.

### Exercício 4 — Método em classe

Na classe `CanalVideo`, adicione um método chamado `mostrarCanal()` que exiba o nome do canal e a quantidade de inscritos.

### Exercício 5 — Criando objetos com classe

Crie dois objetos diferentes a partir da classe `CanalVideo` e mostre os dados usando o método criado.

## Nível Intermediário

### Exercício 6 — Classe `IngressoEvento`

Crie uma classe chamada `IngressoEvento` com:

- `evento`
- `preco`

Adicione o método:

`aplicarDesconto(percentual)`

Esse método deve retornar o novo valor do ingresso com desconto.

### Exercício 7 — Classe `CarteiraDigital`

Crie uma classe chamada `CarteiraDigital` com:

- `titular`
- `saldo`

Adicione os métodos:

- `depositar(valor)`
- `pagar(valor)`

O saldo deve ser atualizado corretamente.

### Exercício 8 — Validação no pagamento

Melhore a classe `CarteiraDigital` para que o método `pagar(valor)` não permita que o saldo fique negativo.

Se o valor for maior que o saldo, exiba uma mensagem de erro.

### Exercício 9 — Classe `SalaVirtual`

Crie uma classe chamada `SalaVirtual` com:

- `nome`
- `capacidade`

Adicione um método chamado `mostrarResumo()` que exiba uma frase com os dados da sala.

### Exercício 10 — Classe com estado

Crie uma classe chamada `LampadaInteligente` com:

- `ligada` (inicialmente `false`)

Adicione os métodos:

- `ligar()`
- `desligar()`

Esses métodos devem alterar o estado da lâmpada.

## Nível Avançado

### Exercício 11 — Herança básica com mídia

Crie uma classe base chamada `Midia` com:

- `titulo`

Adicione um método chamado `reproduzir()`.

Depois, crie uma classe `Musica` que herde de `Midia` e adicione o método `pausar()`. 

### Exercício 12 — Herança com construtor

Crie uma classe base chamada `Dispositivo` com:

- `marca`

Depois, crie uma classe `NotebookGamer` que herde de `Dispositivo` e possua também:

- `placaVideo`

Crie um objeto dessa classe e mostre seus dados.

### Exercício 13 — Sistema de assinaturas

Crie uma classe base chamada `Plano` com:

- `nome`
- `valor`

Depois, crie uma classe `PlanoPremium` que herde de `Plano` e tenha um método chamado `mostrarBeneficios()`.

### Exercício 14 — Sistema de entrega

Crie uma classe base chamada `Entrega` com:

- `codigo`
- `status`

Depois, crie uma classe `EntregaExpressa` que herde de `Entrega` e adicione:

- `prazoHoras`

Inclua um método que mostre um resumo da entrega.

### Exercício 15 — Desafio completo: biblioteca de jogos

Crie uma classe chamada `BibliotecaJogos` que possua:

- uma lista de jogos
- um método `adicionarJogo(nome)`
- um método `listarJogos()`

Depois, teste a classe adicionando alguns jogos e exibindo a lista final.
