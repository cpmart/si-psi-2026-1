\section*{Lista de Exercícios — Objetos Avançados em JavaScript}

# Aula 02 — Objetos Avançados

Conteúdos trabalhados:

- `this`
- desestruturação
- `Object.keys()`
- `Object.values()`
- loops em objetos


## Nível Básico

### Exercício 1 — Perfil de streamer

Crie um objeto chamado `streamer` com as propriedades:

- `canal`
- `plataforma`
- `seguidores`

Adicione um método chamado `mostrarPerfil()` que exiba no console:

`Canal: X | Plataforma: Y | Seguidores: Z`

Use `this` dentro do método.

### Exercício 2 — Corrigindo o acesso com `this`

O código abaixo possui um erro. Corrija-o.

```javascript
const servidor = {
    nome: "Servidor Principal",
    ip: "192.168.0.10",

    mostrarIP: function() {
        console.log(ip)
    }
}
```

### Exercício 3 — Desestruturação simples

Dado o objeto abaixo:

```javascript
const podcast = {
    titulo: "Código Sem Café",
    episodio: 12,
    duracao: 45
}
```

Use desestruturação para criar as variáveis:

- `titulo`
- `duracao`

Depois, mostre ambas no console.

### Exercício 4 — Object.keys()

Dado o objeto:

```javascript
const impressora3D = {
    modelo: "Ender 3",
    material: "PLA",
    status: "imprimindo"
}
```

Mostre no console **todas as chaves** do objeto usando `Object.keys()`. 

### Exercício 5 — Object.values()

Usando o mesmo objeto do exercício anterior, mostre **todos os valores** no console com `Object.values()`. 

## Nível Intermediário

### Exercício 6 — Loop com `for...in`

Crie um objeto chamado `smartwatch` com as propriedades:

- `marca`
- `bateria`
- `conectado`

Use `for...in` para mostrar no console o nome da propriedade e seu valor no formato:

`propriedade: valor`

### Exercício 7 — Desestruturação com renomeação

Dado o objeto:

```javascript
const filme = {
    nome: "Interestelar",
    ano: 2014
}
```

Use desestruturação para criar as variáveis:

- `nomeFilme`
- `anoLancamento`

### Exercício 8 — Somando valores com `Object.values()`

Crie um objeto chamado `pontuacoes` com:

- `fase1`
- `fase2`
- `fase3`

Atribua valores numéricos e use `Object.values()` com um loop para calcular a soma total das pontuações.

### Exercício 9 — Método com `this`

Crie um objeto chamado `playlist` com:

- `nome`
- `quantidadeMusicas`

Adicione um método chamado `resumo()` que mostre uma mensagem como:

`A playlist X possui Y músicas.`

Use `this`.

### Exercício 10 — `Object.entries()`

Dado o objeto:

```javascript
const drone = {
    modelo: "DJI Mini",
    autonomia: 34,
    camera: "4K"
}
```

Use `Object.entries()` para mostrar no console:

- a chave
- o valor

no formato:

`chave: valor`

## Nível Avançado

### Exercício 11 — Verificação de configuração

Crie um objeto chamado `configSistema` com as propriedades:

- `tema`
- `idioma`
- `notificacoes`

Percorra o objeto e verifique se alguma propriedade está vazia (`""`, `null` ou `undefined`).

Se encontrar, exiba:

`Configuração inválida`

### Exercício 12 — Contador de propriedades

Crie uma função chamada `contarPropriedades(obj)` que receba um objeto e retorne quantas propriedades ele possui.

Exemplo esperado:

```javascript
contarPropriedades({a:1, b:2, c:3}) // 3
```

### Exercício 13 — Gerando texto de relatório

Dado o objeto:

```javascript
const sensor = {
    tipo: "temperatura",
    valor: 28,
    unidade: "°C"
}
```

Transforme esse objeto em uma única string no formato:

`tipo=temperatura, valor=28, unidade=°C`

### Exercício 14 — Média de métricas

Crie um objeto chamado `metricasSite` com três valores numéricos, por exemplo:

- `tempoCarregamento`
- `seo`
- `acessibilidade`

Use `Object.values()` para calcular a média dessas métricas.
