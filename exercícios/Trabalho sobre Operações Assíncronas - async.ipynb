# Trabalho Prático – Projeto com APIs (JavaScript – 3º Semestre)

## Objetivo do Trabalho
Desenvolver, em dupla, uma aplicação web utilizando HTML, CSS e JavaScript (lado cliente) que consuma uma ou mais APIs públicas, aplicando os conceitos de:
- fetch
- async/await
- try/catch
- manipulação do DOM
- tratamento de erros
- organização de código

## Contexto
Os sistemas modernos utilizam APIs constantemente. Este projeto simula um cenário real de desenvolvimento onde o aluno consome dados externos, trata erros e apresenta os dados de forma visual.

## Organização
- Trabalho em dupla
- Desenvolvimento e apresentação
- Tempo de apresentação: 8 a 10 minutos

## Requisitos Mínimos (Obrigatórios)
### 1. Requisição à API
- Uso de fetch
- Uso de async/await
- Conversão para JSON

### 2. Tratamento de erro
- Uso de try/catch
- Tratamento de erro de requisição
- Tratamento de dados inválidos

### 3. Interface
- Entrada de dados (input ou botão)
- Atualização dinâmica da página
- Exibição clara das informações (sem console.log())

### 4. Organização do código
- Separação em funções
- Código legível
- Nomes de variáveis claros

### 5. Interação do usuário
- Botão ou campo de busca
- Resposta visual na tela

## Requisitos Desejáveis
- Validação de entrada
- Mensagem de carregamento
- Layout organizado
- Uso de mais de uma API
- Filtros ou busca dinâmica
- Código modular
